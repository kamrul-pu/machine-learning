```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
# Ploting a vector
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units="xy", angles="xy", scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units="xy", angles="xy", scale=1, color="b")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units="xy", angles="xy", scale=1, color="b")
plt.quiver(0,0,-3,-6,scale_units="xy", angles="xy", scale=1, color="r")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# addtion of two vectors
vector_1 = np.asarray([0,0,2,3], dtype="int8")
vector_2 = np.asarray([0,0,3,-2], dtype="int8")

sum_v = vector_1 + vector_2
print(sum_v)
```

```python
plt.quiver(*vector_1,scale_units="xy", angles="xy", scale=1, color="b")
plt.quiver(*vector_2,scale_units="xy", angles="xy", scale=1, color="r")
plt.quiver(*sum_v,scale_units="xy", angles="xy", scale=1, color="g")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Subtraction of two vectors
vector_1 = np.asarray([0,0,2,3], dtype="int8")
vector_2 = np.asarray([0,0,3,-2], dtype="int8")

sub_v = vector_1 - vector_2
print(sub_v)
```

```python
plt.quiver(*vector_1,scale_units="xy", angles="xy", scale=1, color="b")
plt.quiver(*vector_2,scale_units="xy", angles="xy", scale=1, color="r")
plt.quiver(*sub_v,scale_units="xy", angles="xy", scale=1, color="g")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# scalar multiplication.
v1 = np.asarray([0, 0, 2, 3])
v2 = 2 * v1
v2
plt.quiver(*v1,scale_units="xy", angles="xy", scale=1, color="r")
plt.quiver(*v2,scale_units="xy", angles="xy", scale=1, color="b")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# scalar multiplication.
v1 = np.asarray([0, 0, 2, 3])
v2 = -2 * v1
v2
plt.quiver(*v1,scale_units="xy", angles="xy", scale=1, color="b")
plt.quiver(*v2,scale_units="xy", angles="xy", scale=1, color="r")
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
